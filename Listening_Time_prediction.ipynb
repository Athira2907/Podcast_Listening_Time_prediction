```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder,StandardScaler,LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv('train (1).csv')
df2=pd.read_csv('test.csv')
df
```

```python
df.info()
```

```python
# df=df.dropna(subset=['Listening_Time_minutes'])
```

```python
num_col=df.select_dtypes(include=['int64','float64']).columns
num_col
impute=SimpleImputer(strategy='mean')           #strategy--> with what value to replace null
df[num_col]=impute.fit_transform(df[num_col])
```

```python
num=df2.select_dtypes(include=['int64','float64']).columns
num
impute=SimpleImputer(strategy='mean')           #strategy--> with what value to replace null
df2[num]=impute.fit_transform(df2[num])
```

```python
cat_col=df.select_dtypes(include=['object']).columns
num_col
impute=SimpleImputer(strategy='most_frequent')           #strategy--> with what value to replace null
df[cat_col]=impute.fit_transform(df[cat_col])
```

```python
cat=df2.select_dtypes(include=['object']).columns
cat
impute=SimpleImputer(strategy='most_frequent')           #strategy--> with what value to replace null
df2[cat]=impute.fit_transform(df2[cat])
```

```python
XX=df['Podcast_Name']
tf=TfidfVectorizer()
res=tf.fit_transform(XX)
val=pd.DataFrame(res.toarray(),columns=tf.get_feature_names_out())
val
```

```python
XX1=df['Episode_Title']
tf1=TfidfVectorizer()
res1=tf1.fit_transform(XX1)
val1=pd.DataFrame(res1.toarray(),columns=tf1.get_feature_names_out())
val1
```

```python
val1.columns
val1['10']
```

```python
for col in val1.columns:
  val[col]=val1[col]
```

```python
val
```

```python
for col in cat_col[2:]:
  label=LabelEncoder()
  val[col]=label.fit_transform(df[col])
val.head()
```

```python
for col in num_col[1:-1]:
  stand=StandardScaler()
  val[col]=stand.fit_transform(df[[col]])
val.head()
```

```python
x=val
y=df['Listening_Time_minutes']
Xtrain,Xtest,Ytrain,Ytest=train_test_split(x,y,train_size=0.75,random_state=42)
model=LinearRegression()
model.fit(Xtrain,Ytrain)
pred1=model.predict(Xtest)
len(pred1)
```

```python
from sklearn.metrics import r2_score
r2_score(Ytest,pred1)
```
